```python
import pims_nd2
import os
import logging
import labrotation.file_handling as fh
```

```python
logger = logging.getLogger(__name__)
```

```python
root_folder = fh.open_dir("Select root 'tmev' folder!")
```

```python
ca1_mice = ["T301", "T303", "T307", "T314", "T316", "T318", "T324", "T327", "T329", "T330", "T333", "T337", "T347", "T401", "T417", "T426", "T458", ]
nc_mice = ["T349", "T350", "T366", "T391", "T396", "T397",   "T462", "T463", "T531", "T532", "T534"]
mice = ca1_mice + nc_mice
```

# Count all recordings

```python
logging.basicConfig(filename="C:\\Data\\log_test.log", level=logging.DEBUG)
```

```python
dict_imaging_time = {mouse: 0. for mouse in mice}
```

```python
for root, folders, files in os.walk(root_folder):
    for file in files:
        if os.path.splitext(file)[-1] == ".nd2":
            fpath = os.path.join(root, file)
            for mouse in mice:
                if mouse in fpath:
                    nd2_file = pims_nd2.ND2_Reader(fpath)
                    logger.info(fpath)
                    try:
                        framerate = nd2_file.metadata["frame_rate"]
                    except Exception as e:
                        logger.error("\tOSError! Trying with 15 Hz...")
                        framerate = 15.
                    if framerate is not None:  # for stacks, there is no frame rate
                        t_im = nd2_file.sizes["t"]/framerate
                        dict_imaging_time[mouse] += t_im
                    nd2_file.close()
                    break
```

```python
t_total_s = 0.
for key in dict_imaging_time:
    t_total_s += dict_imaging_time[key]
```

```python
t_total_s/(60.*60.)  # convert to h
```

```python
t_ca1_s = 0.
for ca1_mouse in ca1_mice:
    t_ca1_s += dict_imaging_time[ca1_mouse]
```

```python
t_nc_s = 0.
for nc_mouse in nc_mice:
    t_nc_s += dict_imaging_time[nc_mouse]
```

```python
t_ca1_s/3600.
```

```python
t_nc_s/3600.
```

```python
dict_imaging_mice_d7 = {mouse: 0. for mouse in mice}
for root, folders, files in os.walk(root_folder):
    for file in files:
        if os.path.splitext(file)[-1] == ".nd2":
            fpath = os.path.join(root, file)
            include = False
            for expression in expressions_to_look_for:
                if expression in os.path.splitext(fpath)[0]:  # remove ".nd2" which contains d2
                    include = True
                    break
            if include:
                for mouse in mice:
                    if mouse in fpath:
                        print(fpath)
                        nd2_file = pims_nd2.ND2_Reader(fpath)
                        logger.info(fpath)
                        try:
                            framerate = nd2_file.metadata["frame_rate"]
                        except Exception as e:
                            logger.error("\tOSError! Trying with 15 Hz...")
                            framerate = 15.
                        if framerate is not None:  # for stacks, there is no frame rate
                            t_im = nd2_file.sizes["t"]/framerate
                            dict_imaging_mice_d7[mouse] += t_im
                        nd2_file.close()
                        break
```
